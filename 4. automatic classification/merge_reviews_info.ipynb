```python
import pandas as pd
```

```python
df_reviews = pd.read_csv("hotel_reviews.csv")
df_info = pd.read_csv("hotel_info.csv")

display(df_reviews)
display(df_info)
```

```python
df_reviews['hotel_name'].nunique()
```

```python
df_reviews['sentiment_pos'] = df_reviews['sentiment'].apply(lambda x: 1 if x == 1 else 0)
df_reviews['sentiment_neg'] = df_reviews['sentiment'].apply(lambda x: 1 if x == 0 else 0)

display(df_reviews)
```

```python
df_count = df_reviews.groupby("hotel_name").sum().reset_index()[['hotel_name', 'sentiment_pos','sentiment_neg']].rename(columns={'sentiment_pos':'num_of_pos', 'sentiment_neg':'num_of_neg'})

display(df_count)
```

```python
df_merged = df_count.merge(df_info)

display(df_merged)
```
